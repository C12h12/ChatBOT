```python
#the only thing is we are loading the model,keep in mind we are also using same scaler for scaling the input data,you can also save the scaler and then reuse it in differemt code
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import joblib
```

```python
iris=load_iris()
x,y=iris.data,iris.target
```

```python
#data preparation
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)  #the model has already learned the fit values(mean,std) in earlier step,so need to use test data
```

```python
model=LogisticRegression()

#model training
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
score=accuracy_score(y_test,y_pred)
print(score*100,'%')
```

```python
joblib.dump(model,"LR_v1.pkl")
```

```python
#hyperparameter tuning

param_grid={
    'C':[0.1,1.0,10.0],
    'solver':['liblinear','lbfgs']
}

new_models=GridSearchCV(LogisticRegression(),param_grid,cv=5)
new_models.fit(x_train_scaled,y_train)
```

```python
#using the best estimator
best_new_model=new_models.best_estimator_
y_pred=best_new_model.predict(x_test_scaled)

score1=accuracy_score(y_test,y_pred)
print(score1*100,'%')
```

```python
#saving the model
joblib.dump(best_new_model,'LR_v2.pkl')
```

```python
#comparison
print('model 1 accuracy:',score*100,'%')
print('model 2 accuracy:',score1*100,'%')
```

#loading the saved model (LR_v2.pkl)

```python
loaded_model=joblib.load("LR_v2.pkl")
```

```python
input=[[5.1 ,3.5, 1.4, 0.2]]

#scaling the input using same scaler
scaled_input=scaler.transform(input)
```
